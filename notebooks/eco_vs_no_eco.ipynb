```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eco_df = pd.read_csv("C:/Users/Brandon H/Documents/daimil10/cap_stone/sfunding_sat_score/data/sat_data/combined_dallas_data.csv")
```

```python
eco_disadvantaged_df = eco_df[eco_df['Group'] == 'Economically Disadvantaged']
```

```python
eco_disadvantaged_avg_total_by_year = eco_disadvantaged_df.groupby('Year')['Total'].mean()
```

```python
print(eco_disadvantaged_avg_total_by_year)
```

```python

plt.figure(figsize=(10, 6))
plt.plot(eco_disadvantaged_avg_total_by_year.index, eco_disadvantaged_avg_total_by_year.values, marker='o')
plt.title('Average Total Scores for Economically Disadvantaged Students in Dallas County Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Total Score')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
not_eco_disadvantage_df = eco_df[eco_df['Group'] == 'Not Economically Disadvantaged']
```

```python
avg_not_eco_total_by_year = not_eco_disadvantage_df.groupby('Year')['Total'].mean()
```

```python
print(avg_not_eco_total_by_year)
```
