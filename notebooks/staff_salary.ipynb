```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dallas_county_df = pd.read_csv("C:/Users/Brandon H/Documents/daimil10/cap_stone/sfunding_sat_score/data/Staff  Salary Reports/dallas_county_data.csv")
```

```python
filtered_df = dallas_county_df[dallas_county_df['Staff'] == 'TOTAL TEACHING STAFF']

filtered_df = filtered_df[['Year', 'Average Base Pay']]
```

```python
filtered_df['Average Base Pay'] = pd.to_numeric(filtered_df['Average Base Pay'], errors='coerce')
```

```python
avg_base_pay_by_year = filtered_df.groupby('Year')['Average Base Pay'].mean()
```

```python
print(avg_base_pay_by_year)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(avg_base_pay_by_year.index, avg_base_pay_by_year.values, marker='o')
plt.title('Average Base Pay for TOTAL TEACHING STAFF Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Base Pay')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
